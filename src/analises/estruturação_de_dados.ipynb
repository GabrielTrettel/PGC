## Twitter
 
Os arquivos est√£o dispostos em diret√≥rios nomeados pelo `id` do usu√°rio e em seguida os arquivos contendo informa√ß√µes sobre os tweets. O arquivo contendo os dados dos tweets pode segue o seguinte formato:
 
`tweet_id`  |  `tweet_lang`   | `is_RT`

`tweet_content`

```julia
ROOT_DIR = "../../data/tweets/"
TABLE_PATH = "../../../datasets/tweets/tweets.csv"
HEADER = ["user_id", "tweet_id", "tweet_lang", "is_RT", "tweet_content"]
write(TABLE_PATH, join(HEADER, '|')*"\n")

IO = open(TABLE_PATH; append=true)
```

```julia
function save_data(user_id, file_path::String)
    content = readlines(file_path)
    is_RT = false
    
    try
        if content[2][1:2] == "RT"
            is_RT = true
        end
    catch
    end
    
    tweet_id, tweet_lang = split(content[1], "|")[1:end-1]
    
    tweet = replace(content[2], '|' => "\\|")
    
    content = join([user_id, tweet_id, tweet_lang, is_RT, tweet], "|") * "\n"
    
    write(IO, content)
end
```

```julia
for user_folder_id in readdir(ROOT_DIR)
    for tweet_id in readdir(joinpath(ROOT_DIR, user_folder_id))
        if tweet_id[1:4] != "user"
            save_data(user_folder_id, joinpath(ROOT_DIR, user_folder_id, tweet_id))
        end
    end
end
close(IO)
```
